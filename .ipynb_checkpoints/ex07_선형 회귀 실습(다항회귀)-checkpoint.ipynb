```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
boston = pd.DataFrame(data, columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'])
```

```python
boston.head()
```

### Data Scaling
- 모든 특성이 모델에 미치는 영향력을 균형있게 만들어 줌
    - 특정한 특성이 타 특성보다 지나치게 의존하는 것을 방지
- Standard Scaler
    - 평균이 0, 분산 1인 스케일로 변환
- MinMax Scaler
    - 특정 범위(기본적으로 0~1)의 스케일로 변환
    - 이상치에 매우 민감

```python
from sklearn.preprocessing import StandardScaler # 스케일러 도구
```

```python
std_scaler = StandardScaler()

X_scaled = std_scaler.fit_transform(boston)
X_scaled
```

```python
X_scaled = pd.DataFrame(X_scaled, columns = boston.columns) 
# 스케일링을 적용하면, ndarray 형이라 DataFraem으로 다시 형변환!
```

```python
# 데이터 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, 
                                                    target, 
                                                    test_size = 0.3, 
                                                    random_state = 2024)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 1. 일반선형 모델로 예측을 수행해보자!
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score # 교차 검증 도구
```

```python
# 검증으로 성능 확인
result = cross_val_score(LinearRegression(), X_train, y_train, cv = 5)
result.mean()
# r2 score : 데이터에 대해 어느정도 이해하고 있는가를 0~1사이의 값으로 표현된 지표값(1에 가까울수록 우리 모델이 데이터를 잘 이해하고 있다!)
```

#### 회귀평가지표
- MSE(Mean Squared Error, 평균 제곱 오차) : 실제값과 예측값 차이를 제곱해서 평균 낸 것
    - 큰 오차에 더 큰 숫자값을 부여함
- MAE(Mean Absolute Error, 평균 절대 오차) : 실제값과 예측값 차이의 절대값을 평균 낸 것
    - 오차가 선형적으로 부여됨
    - 0~1 사이의 오차값을 가지는 경우 유용하게 활용 가능
- RMSE(Root MSE, 평균 제곱근 오차) : MSE에 루트를 씌운 것
    - 오차의 단위가 원래 단위로 돌아와 해석이 용이함

- R-squared(결정 계수, R^2, r2 score) : 모델이 데이터의 분산(퍼짐 정도)을 얼마나 이해하고 있는가를 나타내는 지표
    - 0 ~ 1 사이의 값으로 표현, 1에 가까울수록 모델이 데이터를 잘 이해하고 있다를 의미 

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score # 회귀 평가 지표 도구
```

```python
# 모델 초기화
lr_model = LinearRegression()
```

```python
# 학습
lr_model.fit(X_train, y_train)
```

```python
# 예측
y_pred = lr_model.predict(X_test)
```

```python
# MSE 값
mean_squared_error(y_test, y_pred)
```

```python
# RMSE 값
mean_squared_error(y_test, y_pred) ** 0.5
```

```python
# MAE 값
mean_absolute_error(y_test, y_pred)
```

```python
# 모델에 내장된 R^2 값
lr_model.score(X_test, y_test)
```

```python
# R^2 값
r2_score(y_test, y_pred)
```

```python
# 2. 경사하강법 모델 평가!
from sklearn.linear_model import SGDRegressor

sgd_model = SGDRegressor(
    eta0 = 0.01, 
    max_iter = 10000, 
    verbose = 1
)

sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)
# Norm : 모델의 가중치 크기
# NNZs : 문제 데이터 개수(0이 아닌 가중치의 개수)
# Bias : 절편값
# T : 가중치 업데이트 횟수
# Avg. loss : 오차값들의 평균
```

```python
print("평균제곱오차 :", mean_squared_error(y_test, y_pred))
print("R^2 :", r2_score(y_test, y_pred))

# 성능이 꽤 괜찮지는 않은 듯 함
# 모델이 단순
```

### 다항회귀 구현하기
1. 수기로 차항 늘리기
2. sklearn 도구 활용(PolynomialFeatures)

```python
# 1. 수기로 차항 늘리기
X_train["LSTATxLSTAT"] = X_train["LSTAT"] ** 2
```

```python
# LSTAT(하위계층의 비율)
plt.figure(figsize = (5, 5))

# 실제값 확인
plt.scatter(X_train["LSTAT"], y_train, label = "original", color = "blue")

# 선형 모델 학습
LSTAT_model = LinearRegression()
LSTAT_model.fit(X_train[["LSTAT", "LSTATxLSTAT"]], y_train)

# 선형 모델의 예측 결과 확인
y_pred = LSTAT_model.predict(X_train[["LSTAT", "LSTATxLSTAT"]])
plt.scatter(X_train["LSTAT"], y_pred, label = "pred", color = "red")

plt.legend()

plt.show()
```

```python
# 2. PolynomialFeatures 특성확장 도구 활용
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 특성확장도구 생성
poly = PolynomialFeatures(2, # 다항식 차수 설정
                         include_bias = False) # 절편을 추가하지 않겠다!
```

```python
# poly 사용을 위해 삭제!
del X_train["LSTATxLSTAT"]
```

```python
# poly에 기준을 위한 데이터 병합본 만들기
X_concat = pd.concat([X_train, X_test], axis = 0)
```

```python
poly.fit(X_concat) # 기준 설정
```

```python
X_train_poly = poly.transform(X_train) # 변환
```

```python
# 기준 잡기 전 컬럼들 확인
poly.feature_names_in_
```

```python
# 확장된 컬럼 담아두기
poly_name = poly.get_feature_names_out(poly.feature_names_in_)
```

```python
# 데이터프레임으로 만들어주기
X_train_poly = pd.DataFrame(X_train_poly, columns=poly_name)
```

```python
# test set도 특성확장시키기!!
X_test_poly = poly.transform(X_test) # 변환
X_test_poly = pd.DataFrame(X_test_poly, columns = poly_name) # 데이터프레임으로 형변환
```

```python
# LinearRegression 모델 학습 및 평가
lr_model = LinearRegression()

lr_model.fit(X_train_poly, y_train)
y_pred = lr_model.predict(X_test_poly)

print(f"MSE : {mean_squared_error(y_test, y_pred)}")
print(f"R^2 : {r2_score(y_test, y_pred)}")

# 특성 확장 전과 후를 비교해보니, 오차값이 줄어드면서 R^2도 올라감! -> 성능이 올라감
# 다항회귀를 통해 더 fit하게 맞춰지는 회귀선(모델)을 만들었기 때문!
```

```python
# SGDRegressor 모델 학습 및 평가
sgd_model = SGDRegressor(
    eta0 = 0.001, 
    verbose = 1, 
    max_iter = 10000
)

sgd_model.fit(X_train_poly, y_train)
y_pred = sgd_model.predict(X_test_poly)

print(f"MSE : {mean_squared_error(y_test, y_pred)}")
print(f"R^2 : {r2_score(y_test, y_pred)}")
```

##### 규제(Regularization)
- w(가중치)가 크다면, X(Feature)가 조금만 달라져도 y는 크게 변함
- w에 억제를 가할 수 있음
---
1. L1 규제 : Lasso
    - 모든 w에 똑같은 힘으로 규제를 적용하는 방법 -> 특정 계수(w)들은 0이 됨
    - 특성 선택(Feature Selection)이 일어남
3. L2 규제 : Ridge
    - 모든 w에 적절한 힘으로 규제를 적용하는 방법
    - 모든 특성을 y를 예측하는데 사용되게끔 규제를 가함(특정 계수들은 0이 되지 않음)

```python
sgd_model = SGDRegressor(
    eta0 = 0.001, 
    verbose = 1, 
    max_iter = 10000, 
    penalty = "l1", # Lasso, 특정 w값이 0이 되게끔 규제를 가하는 방법
    alpha = 0.01 # 규제 강도(알파) : 높이면 높일수록 가하는 규제가 쎄짐
)

sgd_model.fit(X_train_poly, y_train)
y_pred = sgd_model.predict(X_test_poly)

print(f"MSE : {mean_squared_error(y_test, y_pred)}")
print(f"R^2 : {r2_score(y_test, y_pred)}")
```
